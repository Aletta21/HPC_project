```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('temperature_stats.csv')

plt.figure(figsize=(8, 5))
df['mean_temp'].plot(kind='hist', bins=30, edgecolor='black')
plt.title('Distribution of Mean Temperatures')
plt.xlabel('Mean Temperature (°C)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('mean_temp_histogram.png', dpi=150)
```

```python
print(f"Average mean temperature: {df['mean_temp'].mean():.2f} °C")
print(f"Average temperature std deviation: {df['std_temp'].mean():.2f} °C")
print(f"Buildings with >= 50% area > 18°C: {len(df[df['pct_above_18'] >= 50])}")
print(f"Buildings with >= 50% area < 15°C: {len(df[df['pct_below_15'] >= 50])}")
```
